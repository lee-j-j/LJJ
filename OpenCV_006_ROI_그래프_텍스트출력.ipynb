```python
#실습
import cv2
import matplotlib.pyplot as plt

# 이미지 읽기
img = cv2.imread("./images/mario.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지 출력
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

# ROI 영역 자르기
# 배열의 구조 : 배열명[행, 열] → * x:열, y:행
#              행(y)    열(x)
sub_img = img[109:342, 141:353]

plt.figure(figsize=(2,2))
plt.imshow(sub_img)
plt.axis("off")
plt.show()
```

## ROI에 도형 그리기

- <font color=red>cv2.line(배경이미지, 시작점좌표, 끝점좌표, 색상, 선두께)</font> : 직선 그리기

<img src="https://arome1004.cafe24.com/images/opencv/lecture_image/06_line.png" width=40%>  

```python
#실습
img1 = img.copy()

#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

cv2.line(img1, (40, 10), (297, 44), (255, 0, 0), 3)
cv2.line(img1, (82, 453), (278,608), (255, 0, 0), 3)

plt.figure(figsize=(4,4))
plt.imshow(img1)
plt.axis("off")
plt.show()
```

- <font color=red>cv2.circle(배경이미지, 중심좌표, 반지름, 색상, 선두께)</font> : 원 그리기

<img src="https://arome1004.cafe24.com/images/opencv/lecture_image/06_circle.png" width=20%>  

```python
#실습
img2 = img.copy()

#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

# 선 두께를 -1로 하면 채운 원
cv2.circle(img2, (220, 180), 170, (0, 0, 0), 5)
cv2.circle(img2, (220, 180), 10, (255, 0, 0), -1)

plt.figure(figsize=(4,4))
plt.imshow(img2)
plt.axis("off")
plt.show()
```

- <font color=red>cv2.rectangle(배경이미지, 좌상단점좌표, 우하단점좌표, 색상, 선두께)</font> : 사각형 그리기

<img src="https://arome1004.cafe24.com/images/opencv/lecture_image/06_rectangle.png" width=30%>  

```python
#실습
img3 = img.copy()

#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

cv2.rectangle(img3, (141, 109), (353, 342), (0, 255, 255), 5)

plt.figure(figsize=(4,4))
plt.imshow(img3)
plt.axis("off")
plt.show()
```

- <font color=red>cv2.ellipse(배경이미지, 중심좌표, 장축/단축 반지름, 기울기, 호의시작각, 호의 종료각, 색상, 선두께)</font> : 타원 그리기

<img src="https://arome1004.cafe24.com/images/opencv/lecture_image/06_ellipse.png" width=30%>  

```python
#실습
img4 = img.copy()

#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

cv2.ellipse(img4, (239, 256), (220, 260), -80, 90, -90, (0,0,0), 5)

plt.figure(figsize=(4,4))
plt.imshow(img4)
plt.axis("off")
plt.show()
```

- <font color=red>cv2.putText(배경이미지, 출력내용, 출력시작좌표, 폰트, 크기, 색상, 굵기)</font> : 텍스트 그리기

<img src="https://arome1004.cafe24.com/images/opencv/lecture_image/06_text.png" width=30%>  

```python
#실습
img5 = img.copy()

#              x축  y축
# 좌상단 좌표 (141, 109)
# 우하단 좌표 (353, 342)

cv2.putText(img5, "Mario!", (329, 653), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0), 7)

plt.figure(figsize=(4,4))
plt.imshow(img5)
plt.axis("off")
plt.show()
```

## 한글 출력하기

- c:/Windows/Fonts 폴더에서 사용할 폰트를 현재 폴더에 fonts 폴더를 생성하고 복사한다


- <font color=red>font = ImageFont.truetype("fonts/gulim.ttc", 30)</font> : 사용할 폰트와 크기를 가져온다

- <font color=red>img_pil = Image.fromarray(img)</font> : array 이미지를 PIL 이미지로 변환

- <font color=red>draw = ImageDraw.Draw(img_pil)</font> : PIL 이미지를 그림

- <font color=red>draw.text((200, 20), "레나 이미지", font=font, fill=(0, 0, 255, 0))</font> : 이미지에 텍스트를 그림

- <font color=red>img = np.array(img_pil)</font> : PIL 이미지를 numpy 배열로 변환

```python
#실습
import cv2
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import matplotlib.pyplot as plt

img6 = img.copy()

cv2.rectangle(img6, (320, 608), (550, 720), (0, 0, 255), -1)
# 폰트 → 이미지
font = ImageFont.truetype("font/Bazzi.ttf", 70)
# 복사한 이미지 배열을 파이썬 이미지 타입으로 변환
pil = Image.fromarray(img6)
# 이미지를 캔버스에 그리기
draw = ImageDraw.Draw(pil)
# 출력할 텍스트를 캔버스에 그리기
draw.text((329, 633), "마리오~!", font=font, fill=(255, 0, 0, 0))
# 이미지 → 배열
img6 = np.array(pil)
cv2.rectangle(img6, (434, 36), (554, 156), (0, 0, 0), 3)
cv2.ellipse(img6, (239, 256), (220, 260), -80, 90, -90, (0,0,0), 5)
cv2.line(img6, (10, 306), (42, 361), (0, 0, 0), 3)
cv2.line(img6, (88, 453), (180,620), (0, 0, 0), 3)
cv2.line(img6, (174, 571), (180,620), (0, 0, 0), 3)
cv2.line(img6, (140, 595), (180,620), (0, 0, 0), 3)
cv2.circle(img6, (494, 96), 50, (255, 0, 0), -1)
cv2.circle(img6, (494, 96), 25, (255, 255, 0), -1)



plt.figure(figsize=(4,4))
plt.imshow(img6)
plt.axis("off")
plt.show()
```
