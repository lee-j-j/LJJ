```python
import numpy as np
```

```python
# 활성화 함수 (Activation Function) →  계단 함수 (Step Fuction) : 초기에 사용하던 활성화 함수

def activation_func(x) :
  if x <= 0 :
    return 0
  else :
    return 1
```

```python
# 퍼셉트론 함수

def perceptron(x) :
  w = 10
  b = -3
  y = w * x + b
  return activation_func(y)
```

```python
input_data = 10

a1 = perceptron(10)
a2 = perceptron(a1) # like 뉴런 1과 뉴런 2 연결 ≒ 학생점수예측에서 units = 2 한 것
a2
```

### AND 퍼셉트론 만들기

```python
def AND_perceptron(x1, x2) :
  w1 = 1
  w2 = 1
  b = -1
  y = w1*x1 + w2*x2 + b
  return activation_func(y)
```

```python
print(AND_perceptron(1,1)) # 1
print(AND_perceptron(1,0)) # 0
print(AND_perceptron(0,1)) # 0
print(AND_perceptron(0,0)) # 0
```

### OR 퍼셉트론 만들기

```python
def OR_perceptron(x1, x2) :
  w1 = 1
  w2 = 1
  b = 0
  y = w1*x1 + w2*x2 + b
  return activation_func(y)
```

```python
print(OR_perceptron(1,1)) # 1
print(OR_perceptron(1,0)) # 1
print(OR_perceptron(0,1)) # 1
print(OR_perceptron(0,0)) # 0
```

```python
def NAND_perceptron(x1, x2) :
  w1 = -1
  w2 = -1
  b = 2
  y = w1*x1 + w2*x2 + b
  return activation_func(y)
```

```python
print(NAND_perceptron(1,1)) # 0
print(NAND_perceptron(1,0)) # 1
print(NAND_perceptron(0,1)) # 1
print(NAND_perceptron(0,0)) # 1
```

### XOR을 위한 MLP(다층 퍼셉트론)

```python
def XOR_MLP(x1, x2) :
  y = AND_perceptron ((NAND_perceptron(x1,x2)), (OR_perceptron(x1, x2)))
  return y
```

```python
print(XOR_MLP(1,1)) # 0
print(XOR_MLP(1,0)) # 1
print(XOR_MLP(0,1)) # 1
print(XOR_MLP(0,0)) # 0
```

### keras로 XOR MLP 만들기

```python
X = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([0,1,1,0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
```

```python
# 모델 생성

XOR_model = Sequential() # Sequential : 뼈대 역할
XOR_model.add(InputLayer(input_shape=(2,))) # InputLayer : 입력층 (인공신경망에 들어오는 데이터의 형태(input_shape))
XOR_model.add(Dense(units = 4, activation = "sigmoid")) # params : units * 3개 (units 하나당 w1, w2, b 3개)
XOR_model.add(Dense(units = 1, activation = "sigmoid")) # params : units 갯수 + 1 (b)
# 활성화 함수 : 초기에는 계단함수(Step Fuction)를 사용했으나 현재는 sigmoid 사용
```

```python
XOR_model.summary()
```

```python
# 모델 학습
XOR_model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"]) # 분류의 평가 지표 : 정확도(Accuracy)
XOR_model.fit(X, y, epochs = 2000)
```

```python
# 가중치 확인하기
XOR_model.get_layer("dense_8").get_weights() # dense_8 : 중간층  cf) summary
```
